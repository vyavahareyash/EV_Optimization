```python
demand_set=pd.read_csv('Demand_History.csv')
```

```python
supply_set=pd.read_csv('exisiting_EV_infrastructure_2018.csv')
```

```python
mat=np.zeros((4096,100))
for i in range (4096):
    for j in range(100):
        d=demand_set.iloc[i,1:3].values
        s=supply_set.iloc[j,1:3].values
        ans=math.sqrt(np.square(s[0]-d[0])+np.square(s[1]-d[1]))
        mat[i][j]=ans
    
dist=pd.DataFrame(np.transpose(mat))
dist
```

```python
jsum=[]
for j in range(100):
    jsum.append(sum(dist.iloc[j,:].values))
    
```

```python
jsum
```

```python
jsums=sum(jsum)
jsums
```

```python
cost=isums*jsums
cost
```
