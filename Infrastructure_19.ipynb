```python
from pulp import *
import pandas as pd
import numpy as np
```

```python
infra_set=pd.read_excel('infra19lb.xlsx')
infra_set
```

```python
n_supply=100
smax=np.array(infra_set['newsmax19'])
smax
```

```python
model = LpProblem("Supply-infra", LpMinimize)
model
```

```python
index_numbers = [str(i) for i in range(n_supply)]
index_numbers.sort()
```

```python
x_var=[]
for i in range(n_supply):
    lb=int(infra_set.iloc[i:i+1,5].values)
    lb=int(infra_set.iloc[i:i+1,5].values)
    temp=LpVariable('X_{}'.format(index_numbers[i]),cat='Integer', lowBound=lb)
    x_var.append(temp)
x_var=np.array(x_var)
x_var
```

```python
y_var=[]
for i in range(n_supply):
    lb=int(infra_set.iloc[i:i+1,6].values)
    temp=LpVariable('y_{}'.format(index_numbers[i]),cat='Integer',lowBound=lb)
    y_var.append(temp)
y_var=np.array(y_var)
y_var
```

```python
#Cost Equation
cost_eq=lpSum(x_var)+1.5*lpSum(y_var)
print(cost_eq)
```

```python
tot_p=np.array(infra_set['total_parking_slots'])
tot_p
```

```python
model += cost_eq
model
```

for i in range(n_supply):
    lb=int(infra_set.iloc[i:i+1,5].values)
    constraint= x_var[i]>=lb
    model+=constraint
for i in range(n_supply):
    lb=int(infra_set.iloc[i:i+1,6].values)
    constraint= y_var[i]>=lb
    model+=constraint

```python
model
```

```python
#Supply Constraint
for i in range(n_supply):
    constraint = (x_var[i]*200)+(y_var[i]*400) == int(infra_set.iloc[i:i+1,11].values)
    print(constraint)
    model += constraint
```

```python
#Parking Constraint
for i in range(n_supply):
    constraint = x_var[i]+y_var[i] <= tot_p[i]
    print(constraint)
    model += constraint
```

```python
model
```

```python
model.solve(PULP_CBC_CMD())
status =  LpStatus[model.status]
print(status)
```

```python
print("Total Cost:", model.objective.value())
```

```python
p=0
lst=[]
x_var_r=[]
y_var_r=[]
for v in model.variables():
    k=v.value()
    lst.append(int(k))
for i in range(n_supply):
    x_var_r.append(lst[p])
    p+=1
for i in range(n_supply):
    y_var_r.append(lst[p])
    p+=1
x_var_r=np.array(x_var_r)
y_var_r=np.array(y_var_r)
print(x_var_r)
print(y_var_r)
```
