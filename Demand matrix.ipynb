```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
demand=pd.read_csv('Demand_History.csv')
demand
```

```python
demand=demand[['x_coordinate','y_coordinate','2010']]
```

```python
demand
```

```python
tw=demand['2010']
tw
```

```python
mat=np.zeros((64,64))
mat
```

```python
y=0
for i in range(64):
    for j in range(64):
        mat[i][j]=tw[y]
        y+=1
        
mat
```

```python
df=pd.DataFrame(mat)
df
```

```python
df.to_excel('Demand_Matrix.xlsx')
```

```python
import seaborn as sns
```

```python
f = plt.figure()
ax = f.add_subplot(1, 1, 1)
ax.set_xticks(np.arange(65))
ax.set_yticks(np.arange(65))
f.set_figwidth(64)
f.set_figheight(64)
sns.heatmap(df,cmap='crest')
plt.show()
```

```python
x=demand['x_coordinate']
y=demand['y_coordinate']
heatmap,xedges,yedges=np.histogram2d(x,y,bins=50)
extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]]
plt.clf()
plt.imshow(heatmap.T,extent=extent,origin='lower')
plt.show()
```
