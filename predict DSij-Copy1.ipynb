```python
from pulp import *
import pandas as pd
import numpy as np
import math
```

```python
n_supply=50
n_demand=50
demand_set=pd.read_csv('Demand_History.csv')
supply_set=pd.read_csv('exisiting_EV_infrastructure_2018.csv')
```

```python
mat=np.zeros((50,50))
for i in range (50):
    for j in range(50):
        d=demand_set.iloc[j,1:3].values
        s=supply_set.iloc[i,1:3].values
        ans=math.sqrt(np.square(s[0]-d[0])+np.square(s[1]-d[1]))
        mat[i][j]=ans
mat
```

```python
df=pd.DataFrame(mat)
df
```

```python
dist_matrix=mat
dist_matrix
```

```python
ev_demands=np.array(demand_set['2018'].head(50))
```

```python
#variable_names = [str(i)+str(j) for j in range(1, n_demand+1) for i in range(1, n_supply+1)]
variable_names=[str(i) for i in range(50*50)]
variable_names.sort()
variable_names
#print("Variable Indices:", variable_names)
```

```python
DV_variables = LpVariable.matrix("X", variable_names, cat = "Integer", lowBound= 0 )
allocation = np.array(DV_variables).reshape(n_supply,n_demand)
print("Decision Variable/Allocation Matrix: ")
print(allocation)
```

```python
df3=pd.DataFrame(allocation)
df3
```

```python
obj_func = lpSum(allocation*dist_matrix)
print(obj_func)
```

```python
model +=  obj_func
print(model)
```

```python
#Supply Constraints
for i in range(n_supply):
    print(lpSum(allocation[i][j] for j in range(n_demand)) <= supply_matrix[i])
    model += lpSum(allocation[i][j] for j in range(n_demand)) <= supply_matrix[i] , "Supply Constraints " + str(i)
```

```python
# Demand Constraints
for j in range(n_demand):
    print(lpSum(allocation[i][j] for i in range(n_supply)) == ev_demands[j])
    model += lpSum(allocation[i][j] for i in range(n_supply)) == ev_demands[j] , "Demand Constraints " + str(j)
```

```python
#model.solve()
model.solve(PULP_CBC_CMD())

status =  LpStatus[model.status]

print(status)
```

```python
model.writeLP("Supply_demand_prob.lp")
```

```python
print("Total Cost:", model.objective.value())
```

```python
# Decision Variables

for v in model.variables():
    try:
        print(v.name,"=", v.value())
    except:
        print("error couldnt find value")
```
