```python
supply_set = pd.read_csv('exisiting_EV_infrastructure_2019.csv')
demand_set=pd.read_csv('Demand_Predictions.csv')
dist_mat=pd.read_excel('distij.xlsx')
```

```python
demand19=[]
for i in range(4096):
    demand19.append(float(demand_set.iloc[i:i+1,12].values))
demands=dict()
for index,value in enumerate(demand19):
    demands[index]=value
demands
```

```python
names = dist_mat.demand_index.tolist()
others = list(dist_mat.columns)
others.remove('demand_index')         # We don't want "name" to be included

index_tuples = [(name, other) for name in names for other in others]
multi_ix = pd.MultiIndex.from_tuples(index_tuples)
df1 = pd.DataFrame(dist_mat[others].values.ravel(), index=multi_ix, columns=['value'])
dist = df1.to_dict()['value']
dist
```

```python
dist_dict=dict()
p=0
for i in range(4096):
    for j in range(100):
        dist_dict[p]=dist[i,j]
        p+=1
dist_dict        
```
