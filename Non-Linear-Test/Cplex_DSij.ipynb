```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('CPLEXoptimization_solution.csv')
```

```python
df
```

```python
mat=np.zeros((4096,100))
```

```python
lst=df['solution_value']
```

```python
lst
```

```python
p=0
for i in range(4096):
    for j in range(100):
        mat[i][j]=lst[p]
        p+=1
```

```python
mat=pd.DataFrame(mat)
```

```python
mat
```

```python
mat=mat.T
```

```python
mat
```

```python
mat.to_csv('CPLEXoptimization_matrix.csv')
```
