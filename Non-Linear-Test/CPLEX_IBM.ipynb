```python
i = 100
j = 4096
set_D = range(j)
set_S = range(i)
```

```python
supply_set = pd.read_csv('exisiting_EV_infrastructure_2018.csv')
demand_set=pd.read_csv('Demand_Predictions.csv')
dist_mat=pd.read_csv('Distance_Matrix.csv')
```

```python
df1 = pd.DataFrame(dist_mat[dist_mat.columns].values.ravel(), index=multi_ix, columns=['value'])
```

```python
dist = df1.to_dict()['value']
dist
```

```python
import docplex.mp.model as cpx
opt_model = cpx.Model(name="Model")
```

```python
# if x is Continuous
x_vars  = {(i,j): opt_model.continuous_var(lb=0,name="x_{0}_{1}".format(i,j)) for i in set_S for j in set_D}
```

```python
s_lst=supply_set['existing_num_SCS']
f_lst=supply_set['existing_num_FCS']
```

```python
# if S is Continuous
S_vars  = {i: opt_model.continuous_var(lb=s_lst[i],name="S_{0}".format(i)) for i in set_S}
```

```python
# if F is Continuous
F_vars  = {i: opt_model.continuous_var(lb=f_lst[i],name="F_{0}".format(i)) for i in set_S}
```

```python
# Supply constraints
constraints = {i : opt_model.add_constraint(ct=opt_model.sum(x_vars[i,j] for j in set_D) <= (200*S_vars[i]+400*F_vars[i]),
                                            ctname="constraint_capacity_{0}".format(i)) for i in set_S}
```

```python
# Dwmand constraints
constraints = {j : opt_model.add_constraint(ct=opt_model.sum(x_vars[i,j] for i in set_S) >= demands[j]-0.001,
                                            ctname="Demands_{0}".format(j)) for j in set_D}
```

```python
park_lst=supply_set['total_parking_slots']-supply_set['existing_num_SCS']-supply_set['existing_num_FCS']
```

```python
#Parking Constraint
constraints = {i : opt_model.add_constraint(ct= S_vars[i]+F_vars[i] <= park_lst[i],
                                            ctname="park_{0}".format(i)) for i in set_S}
```

```python
objective = opt_model.sum(x_vars[i,j] * dist[i,j] 
                          for i in set_S 
                          for j in set_D)
```

opt_model.export_as_lp('cplex_model.lp')

```python
# for minimization
opt_model.minimize(objective)
```

```python
# solving with local cplex
opt_model.solve()
```
