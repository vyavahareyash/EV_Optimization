```python
mat=np.zeros((4096,100))
for i in range (4096):
    for j in range(100):
        d=demand_set.iloc[i,1:3].values
        s=supply_set.iloc[j,1:3].values
        ans=math.sqrt(np.square(s[0]-d[0])+np.square(s[1]-d[1]))
        mat[i][j]=ans
    

    
```

```python
mat
```

```python
df=pd.DataFrame(np.transpose(mat))
```

```python
df
```

```python
df.to_excel('distijtrans.xlsx')
```
