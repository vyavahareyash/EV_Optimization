```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
demand_set=pd.read_csv('Demand_History.csv')
```

```python
demand_set
```

```python
x = np.array([[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017]])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg3=PolynomialFeatures(degree=3)
x_poly_3=poly_reg3.fit_transform(x)
```

```python
x_poly_3
```

```python
ncol1=[]
for i in range(4096):
    y = demand_set.iloc[i,3:-1].values
    lin_reg_3=LinearRegression()
    lin_reg_3.fit(x_poly_3,y)
    ans1=lin_reg_3.predict(poly_reg3.transform([[2018]]))
    ncol1.append(ans1[0])
```

```python
demand_set['2018p']=ncol1
```

```python
demand_set
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(demand_set['2018'],demand_set['2018p']))
```

```python
from sklearn.metrics import r2_score
r2_score(demand_set['2018'],demand_set['2018p'])
```
