```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dset=pd.read_csv('Demand_History.csv')
sset=pd.read_csv('exisiting_EV_infrastructure_2018.csv')
```

```python
dset
```

```python
sset
```

```python
supply=sset.iloc[:,1:3].values
demand=dset.iloc[:,1:3].values
```

```python
supply
```

```python
demand
```

```python
xd=demand[:,0]
yd=demand[:,1]
xs=supply[:,0]
ys=supply[:,1]
```

```python
f = plt.figure()
ax = f.add_subplot(1, 1, 1)
ax.set_xticks(np.arange(65))
ax.set_yticks(np.arange(65))
f.set_figwidth(64)
f.set_figheight(64)
plt.grid()
plt.scatter(xd, yd, color='red',s=200)
plt.scatter(xs,ys,color='green',marker='*',s=700)
plt.legend(['Demand points','Supply points'],prop={'size': 30})
plt.title('64x64 Grid Map', fontdict={'fontsize': 50})
plt.show()
```
