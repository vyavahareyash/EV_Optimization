```python
import numpy as np
import pandas as pd
```

```python
demand_history=pd.read_csv('Demand_History.csv')
demand_history
```

```python
n_demand=demand_history.shape[0]
```

```python
x1 = np.array([[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018]])
x2 = np.array([[2010],[2011],[2012],[2013],[2014],[2015],[2016],[2017],[2018],[2019]])
print(x1)
print(x2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
lin_reg=LinearRegression()
```

```python
poly_reg=PolynomialFeatures(degree=3)
x_poly1=poly_reg.fit_transform(x1)
x_poly2=poly_reg.fit_transform(x2)
print(x_poly1)
print(x_poly2)
```

```python
y_test1=poly_reg.transform([[2019]])
y_test2=poly_reg.transform([[2020]])
print(y_test1)
print(y_test2)
```

```python
col19=[]
for i in range(n_demand):
    y = demand_history.iloc[i,3:].values
    lin_reg.fit(x_poly1,y)
    ans=lin_reg.predict(y_test1)
    col19.append(ans[0])
```

```python
demand_history['2019']=col19
demand_history
```

```python
col20=[]
for i in range(n_demand):
    y = demand_history.iloc[i,3:].values
    lin_reg.fit(x_poly2,y)
    ans=lin_reg.predict(y_test2)
    col20.append(ans[0])
```

```python
demand_history['2020']=col20
demand_history
```

demand_history.to_csv('Demand_Predictions.csv',index=False)
