```python
import pandas as pd
import numpy as np
import math
```

```python
n_supply=100
n_demand=4096
demand_set=pd.read_csv('Demand_History.csv')
supply_set=pd.read_csv('exisiting_EV_infrastructure_2018.csv')
```

```python
mat=np.zeros((n_supply,n_demand))
for i in range (n_supply):
    for j in range(n_demand):
        d=demand_set.iloc[j,1:3].values
        s=supply_set.iloc[i,1:3].values
        ans=math.sqrt(np.square(s[0]-d[0])+np.square(s[1]-d[1]))
        mat[i][j]=ans
mat
```

```python
dist_matrix=pd.DataFrame(mat)
dist_matrix
```

dist_matrix.to_csv('Distance_Matrix.csv',index=False)
